## Exploring and Visualizing a Simple Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Iris dataset
iris_df = sns.load_dataset('iris')
print("Dataset loaded successfully!")
```

### Inspecting the Dataset

```python
print("Shape of the dataset:", iris_df.shape)
```

```python
print("\nColumn names:", iris_df.columns.tolist())
```

```python
print("\nFirst 5 rows of the dataset:")
display(iris_df.head())
```

```python
print("\nDataset Info:")
iris_df.info()
```

```python
print("\nDescriptive statistics:")
display(iris_df.describe())
```

### Visualizing the Dataset

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=iris_df,
    x='petal_length',
    y='petal_width',
    hue='species',
    s=100, # size of the points
    alpha=0.7 # transparency
)
plt.title('Petal Length vs. Petal Width of Iris Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Species')
plt.show()
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=iris_df,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    s=100,
    alpha=0.7
)
plt.title('Sepal Length vs. Sepal Width of Iris Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Species')
plt.show()
```

### Histograms for Value Distributions

```python
# Create histograms
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris_df[col], kde=True, bins=10)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel(f'{col} (cm)')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Box Plots for Outlier Identification

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=iris_df[col], x='species', hue='species', data=iris_df, palette='viridis', legend=False)
    plt.title(f'Box Plot of {col.replace("_", " ").title()} by Species')
    plt.xlabel('Species')
    plt.ylabel(f'{col} (cm)')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
